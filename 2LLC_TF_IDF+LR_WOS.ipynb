```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
import time
```

```python
train = pd.read_csv('/content/drive/My Drive/HTC_CNN_dataset/WOS/wos_train_final.csv')
test = pd.read_csv('/content/drive/My Drive/HTC_CNN_dataset/WOS/wos_test_final.csv')
```

```python
train_data = train
test_data = test

tfidf = TfidfVectorizer()
```

```python
train.head()
```

```python
test.head()
```

```python
X_train_text = train_data['text'].values
y_train_l1 = train_data['l1'].values
y_train_l2 = train_data['l2'].values

X_test_text = test_data['text'].values
y_test_l1 = test_data['l1'].values
y_test_l2 = test_data['l2'].values
```

```python
start_time = time.time()
model_l1 = make_pipeline(TfidfVectorizer(), LogisticRegression())
model_l1.fit(X_train_text, y_train_l1)
end_time = time.time()
time_l1 = round(end_time - start_time)
print(f'Time for l1: {time_l1}')

accuracy_l1 = accuracy_score(y_train_l1, model_l1.predict(X_train_text))
print(f'Level 1 Accuracy (Training Set): {accuracy_l1:.4f}')
```

```python
start_time = time.time()
model_l2 = make_pipeline(TfidfVectorizer(), LogisticRegression())
model_l2.fit(X_train_text, y_train_l2)
end_time = time.time()
time_l2 = round(end_time - start_time)
print(f'Time for l2: {time_l2}')

accuracy_l2 = accuracy_score(y_train_l2, model_l2.predict(X_train_text))
print(f'Level 2 Accuracy (Training Set): {accuracy_l2:.4f}')
```

```python
test_pred_l1 = model_l1.predict(X_test_text)
test_pred_l2 = model_l2.predict(X_test_text)
```

```python
test_accuracy_l1 = accuracy_score(y_test_l1, test_pred_l1)
test_accuracy_l2 = accuracy_score(y_test_l2, test_pred_l2)
```

```python
print(f'Test Accuracy for l1: {test_accuracy_l1:.4f}')
print(f'Test Accuracy for l2: {test_accuracy_l2:.4f}')
```

```python
import numpy as np

correct_l1 = test_pred_l1 == y_test_l1
accuracy_l1 = np.mean(correct_l1)
print(f"Consistency Accuracy for l1: {accuracy_l1:.4f}")


correct_l1_indices = np.where(correct_l1)[0]
pred_labels_l2_filtered = test_pred_l2[correct_l1_indices]
labels_l2_filtered = y_test_l2[correct_l1_indices]

correct_l2 = pred_labels_l2_filtered == labels_l2_filtered
accuracy_l2 = np.sum(correct_l2) / len(y_test_l2)
print(f"Consistency Accuracy for l2: {accuracy_l2:.4f}")
```
